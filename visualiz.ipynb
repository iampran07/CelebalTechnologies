```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('Iris.csv')
```

```python
iris.head()
```

```python
plt.figure(figsize=(8, 6))
colors = {'Iris-setosa': 'red', 'Iris-versicolor': 'green', 'Iris-virginica': 'blue'}
plt.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'], c=iris['Species'].map(colors))
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
iris.hist(edgecolor='black', linewidth=1.2, figsize=(12, 8))
plt.suptitle("Histograms of Iris Features", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
iris.boxplot()
plt.title("Boxplot of Iris Features")
plt.show()
```

```python
means = iris.groupby('Species').mean()
features = means.columns

for feature in features:
    means[feature].plot(kind='bar', figsize=(8, 6), title=f'Mean {feature.capitalize()} by Species', legend=True)
    plt.ylabel('Mean Value')
    plt.xlabel('Species')
    plt.xticks(rotation=0)
    plt.show()
```
